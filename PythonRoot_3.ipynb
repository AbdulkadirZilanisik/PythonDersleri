# split metodu

---




```python
#split metodu spliten sonra gelen parantezde belirtilen yerden cümleleri yada kelimelerli ayırır
adi="Abddulkadir Zilan Işık"
print(adi)
print(adi.split(" "))
adi=adi.split(" ")
Adi=adi[2]
print(Adi)
ADi=adi[1]
print(ADi)
```

upper metodu


```python
#upper değişkenin bütün harflerini büyültür
adi="Abddulkadir Zilan Işık"
print(adi)
print(adi.upper())
```

lower metodu

```python
#lower değişkenin bütün harflerini küçültür
adi="ABDDULKADIR ZILAN IŞIK"
print(adi)
print(adi.lower())
```

title metodu

```python
# title değişkenin bütün kelimelerin ilk harfini büyük yapar
adi="ABDDULKADIR ZILAN IŞIK"
print(adi)
print(adi.title())
```

capitalize

```python
#capitalize str değişkenin sadece ilk kelimenin sadece baş harfini büyük yazar
adi="ABDDULKADIR ZILAN IŞIK"
print(adi)
print(adi.capitalize())
```

strip metodud

```python
#strip metodu değişkenin başındaki ve sonundaki boşlukları siler
adi="****ZILAN IŞIK****"
print(adi)
print(adi.strip("*"))
print(adi.rstrip())
print(adi.lstrip())
adi="_______Abdulkadir Zilan Işık_____"
print(adi)
print(adi.strip("_"))
# strip sonra gelen parantezde ne belirtilirse onları değişkenin başından ve sonundan siler
#rstrip sonundakileri siler
#lstrip başındakileri siler
print(adi.rstrip("_"))
print(adi.lstrip("_"))
```

replace metodu

```python
# replace metodu replace sonra gelen parntezde belirtilen ilk özellik yerine ikinci özelliği yazar
metin="batman.edu.tr"
print(metin)
print(metin.replace("batman.edu.tr","https://batman.edu.tr"))
```

find metodu

```python
# find metodu belirtilen özelliğin yada kelimenin nerde başladığını yazar
metin="my name is hasan she name cemre he name murat"
print(metin.find("name"))
print(metin.find("h"))
print(metin.find("name",10,30))
print(metin.find("z"))
```

index metodu

```python
#index metodu find metodu ile aynı işi yapar
metin="my name is hasan she name cemre he name murat"
print(metin.index("name"))
print(metin.index("h"))
print(metin.index("name",10,30))
```

count metodu

```python
# count belirtilen özelliğin kaçte olduğunu yazdırır
adi="Batman Üniversitesi"
print(adi.count("a"))
print(adi.count("a",5,10))
```

swith metodu

```python
# startswith değişken belirtilen özellikle başlıyo ise true başlamıyo ise false yazdırır
adi="Abdulkadir zilan ışık"
print(adi.startswith("Abdulkadir"))
print(adi.startswith("abdulkadir"))
#endswith ise değişken belirtilen özelik ile bitiyomu diye bakar
print(adi.endswith("ışık"))
print(adi.endswith("ışıK"))


```

isalpha isdigit metodları


```python
# isalpha değişkendeki tüm karakterler alfabetik mi diye sorar
# isdigit değişkendeki tüm karakterle sayı mı diye sorar
adi="Abdulkadir zilan ışık"
alfabetik="abcd"
sayi="10"
print(adi.isalpha())
print(alfabetik.isalpha())
print(adi.isdigit())
print(sayi.isdigit())
```

#kelimenin yerini belirtme metodu-kelimlerin sayısını da dahil eder

```python
adi="Batman"
print(adi.center(12,"*"))
#center kelimeyi belirtilen uzunlukta ortalar ve sonra başına ve sonun belirtilen özelliği koyar
print(adi.ljust(25,"*"))
print(adi.rjust(25,"*"))
```
